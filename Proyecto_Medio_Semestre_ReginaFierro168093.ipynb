1. ¿Quién falta más seguido, los hombres o las mujeres?

```python
sns.catplot(
    data=df, x="age", y="absences", col="sex",
    kind="bar", height=4, aspect=1,
)
plt.axhline(y=3, xmin=0, xmax=1.5)
plt.style.use("dark_background") 
```

Se hizo un análisis de ausencias entre ambos sexos y las edades, se puede ver que los hombre faltan más que las mujeres. Categorizando ambos sexos, en el caso de los hombres los estudiantes que suelen tener más ausencias son los estudiantes de 22 años y los que tienen menos ausencias son los de 15 años.
En el caso de las mujeres las estudiantes que tienen mayor ausencia son las de 19 años y las de menor ausencia de 15 años.
