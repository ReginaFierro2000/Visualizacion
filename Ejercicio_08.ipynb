### 2. Modifica el parámetro _text_ para que cada punto tenga su mes indicado.

```python
dj = sns.load_dataset("dowjones")
pd.to_datetime(dj['Date'])
#dj['year'] = pd.DatetimeIndex(dj['Date']).year
dj['month'] = pd.DatetimeIndex(dj['Date']).month
dj.head(100)
```

```python
fig = px.scatter(dj, x="month", y="Price", hover_data = ["Date"], color = "Price") #, size = 'Price',color = 'Date',title='El aumento de precios en 1965')
fig.show()
```

### 3. Repite el gráfico anterior, pero agrega _markers_

```python
fig = px.scatter(dj, x="month", y="Price", hover_data = ["Date"], color = "Price") #, size = 'Price',color = 'Date',title='El aumento de precios en 1965')

fig.update_traces(marker=dict(size=12,
                              line=dict(width=2,
                                        color='DarkSlateGrey')),
                  selector=dict(mode='markers'))
fig.show()
```

### 4. Ahora, agrega un gráfico de línea con datos de 1960 en adelante. Usa el zoom de tu gráfico interactivo para ver el comportamiento en esta década.

```python
dj = sns.load_dataset("dowjones")
pd.to_datetime(dj['Date'])
#dj['year'] = pd.DatetimeIndex(dj['Date']).year
#mask = (dj[(dj['year'] > 1960) & (dj['year'] < 1968)])
filtered_df =dj.loc[dj["Date"].between('1960-01-01', '1968-12-01')]
print(filtered_df)
```
