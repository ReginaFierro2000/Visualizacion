```python
total_G3  = df.G3.sum()
print('Total de los estudiantes: %s' % total_G3)
```

```python
# compute percentage of each format
percentage = []
 
for i in range(df.shape[0]):
    pct = (df.G3[i] / total_G3) * 100
    percentage.append(round(pct, 2))
 
# display percentage
print(percentage)
 
# display data
df['Percentage'] = percentage
#display(df)
df['Percentage'].count()
```

```python
# compute percentage of each format
percentage = []
for i in range(df.shape[0]):
    pct = (df.G3[i] / total_G3) * 100
    percentage.append(round(pct,2))
df['Percentage'] = percentage
 
# depict illustration
plt.figure(figsize=(8,25))
colors_list = ['Red','Orange'] #, 'Blue', 'Purple', 'pink','yellow','green']
graph = plt.bar(df.sex,df.G3, color = colors_list)
plt.title('Percentage of runs scored by MS Dhoni across all formats')
 
i = 0
for p in graph:
    width = p.get_width()
    height = p.get_height()
    x, y = p.get_xy()
    plt.text(x+width/2,
             y+height*1.01,
             str(df.Percentage[i])+'%',
             ha='center',
             weight='bold')
    i+=1
plt.show()
```
