# 2. Principios de diseño de Informacion

¿Cómo percibimos la información? ¿Y cómo, siendo creadores de visualizaciones, podemos usar estas percepciones para comunicar de manera más efectiva nuestros datos? 

La __teoría de la Gestalt__ (union, en Aleman) es una de las formas en que podemos
pensar en cómo nuestros lectores verán nuestros gráficos. Esta se desarrolló en el principios del siglo XX por psicólogos alemanes y se refiere a cómo tendemos a Organizar los elementos visuales en grupos. En vez de pensar en elementos individuales, tendemos a buscar como agrupar la informacion en un todo.

Otros desarrollos en el campo fueron interrumpidos por
el surgimiento del régimen nazi en Alemania y luego por la Segunda Guerra Mundial, y después de la guerra fue
criticado por no tener métodos metodológicos rigurosos. Pero las ideas persisten.

Algunos de los principios mas importantes a tomar en cuenta para el diseño de Informacion son los siguientes:

## I) Simplicidad

Nuestra mente percibe todo en su forma más simple. 

Por ejemplo, al analizar la siguiente imagen vemos como los elementos individuales no tienen significado alguno. Sin embargo, juntos, nuestra mente los agrupa para deletrear la palabra ‘logo’.

Screen Shot 2023-01-26 at 17.10.41.png

## II) Proximidad

Tendemos a percibir los objetos cercanos entre si, como parte de lo mismo. Estop aplica tanto paa texto u otros elementos cerca entre si, como una serie de puntos que estan muy cerca entre si.

```python
data = sns.load_dataset('iris')
```

```python
plt.scatter(data["petal_length"], data["petal_width"])
plt.show()
```

## III) Similaridad

Nuestros cerebros tienden a agrupar entre si elementos que comparten colores, formas, etc. Con el ejemplo anterior:

```python
plt.scatter(data.loc[data["species"] == "setosa", "petal_length"], 
            data.loc[data["species"] == "setosa", "petal_width"], marker="o")
plt.scatter(data.loc[data["species"] == "versicolor", "petal_length"], 
            data.loc[data["species"] == "versicolor", "petal_width"], marker="v")
plt.scatter(data.loc[data["species"] == "virginica", "petal_length"], 
            data.loc[data["species"] == "virginica", "petal_width"], marker="s")
plt.show()
```

## IV) Cierre

No nos gustan los espacios incompletos: tendemos a cerrar esos espacios de la forma mas simple posible, aun si eso no concuerda del todo con la realidad.

Screen Shot 2023-01-26 at 16.39.51.png

En la imagen de la derecha, por ejemplo, alguno podria irse con la finta y creer que los datos faltantes corresponden a una sola recta, aun si es mas logico que los valores primero suban y luego bajen, como en el resto de los datos.

## V) Continuidad

Los objetos que están alineados o se continúan entre sí se perciben como un grupo.
Por lo tanto, nuestros ojos siguen el camino mas natural para ellos. No necesitas la línea del eje horizontal en este gráfico de barras, por ejemplo, porque las barras están alineadas a lo largo de un
camino consistente.

## VI) Procesamiento preatentivo

*Este proceso es, quiza, el mas importante al hacer buenas visualizaciones.*


Nuestra mente es capaz de detectar anomalias mas facilmente cuando tenemos una guia para ello.

Trata, por ejemplo, de encontrar el valor mas grande en la siguiente tabla:

```python
df2 = pd.read_csv("https://gist.githubusercontent.com/armgilles/194bcff35001e7eb53a2a8b441e8b2c6/raw/92200bc0a673d5ce2110aaad4544ed6c4010f687/pokemon.csv")

grouped = df2.groupby("Type 1")[["HP", "Attack", "Defense", "Sp. Atk", "Sp. Def", "Speed"]].mean()
grouped
```

Dificil, ¿no?

Ahora, te presento la misma tabla, pero con una ligera modificacion. Vuelve a tratar de encontrar el valor mas alto de toda la tabla.

```python
grouped.style.highlight_max(axis=None)
```
